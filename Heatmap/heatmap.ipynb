```python
import pandas as pd
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Import data
data = pd.read_csv('data.csv')
```

```python
#Calculating correlation coefficients using pandas' corr function
corr = data.corr()
```

```python
import matplotlib.colors as mcolors
cmap = mcolors.LinearSegmentedColormap.from_list('custom_colors', 
    [(235/255,111/255,93/255),
     (113/255,201/255,221/255),
     (51/255,179/255,159/255),
     (99/255,118/255,160/255),
     (245/255,175/255,152/255)], N=200)
```

```python
# Creating a heat map

# Setting the graphic size 
plt.rcParams['figure.figsize'] = 9,8
# Setting the font
plt.rcParams['font.family'] = 'Times New Roman'
# Setting seaborn global parameter  
sns.set(color_codes=True,font_scale=1.2,font = 'Times New Roman')    #seaborn  ()

#heatmap
ax = sns.heatmap(corr,vmin=-1,vmax=1,center=0,annot=True,annot_kws={'size':9},fmt='.2f',
                 cmap = sns.diverging_palette(20,200,n=200),cbar_kws={'shrink':.7,'format':"%.1f"},
                 linewidths=.5,square=True)

ax.set_yticklabels(ax.get_xticklabels(),size=10)

# Place the X-axis label on top
ax.xaxis.set_ticks_position("top")
# Do not show axes
ax.xaxis.set_ticks_position("none")

ax.set_xticklabels(ax.get_xticklabels(),rotation = 90,size=10)
```
